```python
import pandas as pd
import matplotlib
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
complete_dataset = pd.read_csv('CompleteDataset.csv')
```

```python
complete_dataset.head()
```

```python
complete_dataset.boxplot(column="Overall",        # Column to plot
                 by= "RCM",         # Column to split upon
                 figsize= (8,8))
```

```python
complete_dataset.plot(kind='hexbin', x='Overall', y='RCM', gridsize=20)
```

```python

overall = np.array(complete_dataset['Overall'])
overall.sort()
```

```python
rcm = np.array(complete_dataset['RCM'])
rcm.sort()
```

```python
X , Y = np.meshgrid(overall, rcm)
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)
plt.contour(X,Y,Z, colors='green');
```
